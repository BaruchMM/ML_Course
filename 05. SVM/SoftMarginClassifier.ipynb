```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
import matplotlib.cm as cm
```

```python
# 2 point example
X = [[-2,2],[3,-1]]
y = [1,2]

# 3 point example
X = [[-2,2],[3,-1],[-4,-2]]
y = [1,2,1]

# n point example
n = 100
X, y = make_blobs(n,2,centers=2, random_state=0)
```

```python
# Using linear kernel
clf = SVC(kernel = "linear", C=1)
# Fit to data
clf.fit(X,y)

# Fit info
verbose = True
if verbose:
  # print("Decision function: " + str(clf.decision_function(X)))
  print("Prediction: " + str(clf.predict(X)))
  print("Support Vectors: " + str(clf.support_vectors_))
  print("Lagrange Multipliers: " + str(np.abs(clf.dual_coef_)))
```

```python
# Plotting the predictions
# the ones with x where not well classified
y_predicted = clf.predict(X)

fig, ax = plt.subplots(figsize=(8,5), dpi=100)

for i in np.unique(y):
  Xs = X[y==i]
  plt.scatter(Xs.T[0],Xs.T[1], alpha=0.2)

for i in np.unique(y_predicted):
  Xs = X[y_predicted == y]
  y_predicteds = y_predicted[y_predicted == y]
  Xss = Xs[y_predicteds==i]
  plt.scatter(Xss.T[0],Xss.T[1], marker=".")

for i in np.unique(y_predicted):
  Xs = X[y_predicted != y]
  y_predicteds = y_predicted[y_predicted != y]
  Xss = Xs[y_predicteds==i]
  plt.scatter(Xss.T[0],Xss.T[1], marker="x")


ax.set_title("SV classifier linear kernel")
```
