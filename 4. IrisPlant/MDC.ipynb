```python
import numpy as np
import matplotlib.pyplot as plt

import myscripts

n_centroids = 5
dim = 2
# generate random centroids 
centroids = np.array(np.random.random((n_centroids,dim))*2-1)

npt_blobs = 100

# blobs arround centroids
pts = np.zeros(len(centroids.T))
for i in range(len(centroids)):
    pts = np.vstack((pts,np.random.normal(centroids[i],[0.2]*dim,(npt_blobs,len(centroids[0])))))
pts = pts[1:]

true_indexes = np.ndarray.flatten(np.array([[i]*npt_blobs for i in range(n_centroids)]))

# scatter the points
myscripts.cluster_scatter(pts,true_indexes)

indexes = myscripts.MDC(pts,centroids).indexes

```

```python
myscripts.cluster_scatter(pts,indexes)
```

```python
percentage = np.sum(true_indexes == indexes)/(npt_blobs*n_centroids)*100
print(f"MDC chose {percentage}% of total points correctly.")
```
