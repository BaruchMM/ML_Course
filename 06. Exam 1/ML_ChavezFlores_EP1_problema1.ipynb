# Ejercicio 1

El dataset1 adjunto contiene 600 puntos en ℜ2. Encuentre los centros de los cinco grupos
en los cuáles se distribuyen estos 600 puntos utilizando el método k-means. Grafique los
puntos originales y los centros encontrados con k -means. Guarde su gráfica con puntos
originales y centros encontrados en un archivo llamado ML_Apellidos_EP1_resulados.PDF (2
puntos).


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster
```

```python
# Importamos los datos
datos = np.genfromtxt('dataset1.csv', delimiter=',')
```

```python
# Usamos kmeans de sklearn
k = 5
kmeans = cluster.k_means(datos,k)

# Guardamos los centroides y etiquetas en arreglos separados
centroides = kmeans[0]
etiquetas = kmeans[1]

# Graficamos
fig = plt.figure(figsize=(8,5), dpi = 120)
# cada etiqueta será graficada con un color
for i in range(len(np.unique(etiquetas))):
  datos_graf = datos[etiquetas == i]
  plt.scatter(datos_graf.T[0], datos_graf.T[1])
# graficamos los centroides en color negro
plt.scatter(centroides.T[0], centroides.T[1],color = "k", label="Centroides")
plt.legend()
```
