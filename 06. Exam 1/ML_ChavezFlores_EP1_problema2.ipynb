Utilice el método PCA para encontrar las primeras 2 componentes principales de los datos proporcionados en el dataset2. Grafique un scatterplot con la información de las 2 componentes principales encontradas . Añada el scatterplot a su archivo ML_Apellidos_EP1_resulados.PDF (2 puntos).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import decomposition
```

```python
# Importamos los datos
datos = np.genfromtxt('dataset2.csv', delimiter=',')
```

```python
# Definimos el modelo de PCA de sklearn
miPCA = decomposition.PCA(2)
miPCA.fit(datos)

# Calculamos la proyección de los datos
proyeccion = miPCA.transform(datos)

# Graficamos la proyección
fig = plt.figure(figsize=(8,5), dpi = 120)
plt.scatter(proyeccion.T[0],proyeccion.T[1])

```
